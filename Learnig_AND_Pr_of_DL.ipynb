Breast Cancer Classification with a Simple Neural Network (NN)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection & Processing

```python
# loading the data from skleaern
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
 #Loading the data to a dataframe
 data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
#Adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
#print last 5 rows of the dataframe
data_frame.tail()
```

```python
#Number of rows and columns in the dataset
data_frame.shape
```

```python
#getting some information about the data
data_frame.info()
```

```python
#statistical measures about the data
data_frame.describe()
```

Separating the features and target

when u will drop the columns we will mention axis=1                  or deop the columns  we will mention axis=0                          

```python
X = data_frame.drop(columns= 'label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print (X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
from sklearn import StandardScaler # type: ignore
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

 **The** **Neural** **Network**

```python
# importing tenserflow and keras
import tensorflow as tf 
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural network
#sigmoid is Logistic Regrassion fun.

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),                             #input layer
    keras.layers.Dense(20, activation='relu'),                             #hidden layer
    keras.layers.Dense(2, activation='sigmoid')                          #output layer
])
```

```python
# compiling the neural network

model.compile(optimizer='adam',
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])
```

```python
# training the Meural nrtwork

history = model.fit(X_train_std, Y_train, validation_split=0.1,  epochs=10)
```

Visualization Accuracy And Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='upper right')
# plt.show()
```

Accuracy of the Model On the test Data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
